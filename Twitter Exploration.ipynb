#Problem 2: Analyzing Tweets and Tweet Entities with Frequency Analysis

**1. Word Count:** 
* Use the tweets you collected in Problem 1, and compute the frequencies of the words being used in these tweets. 
* Plot a table of the top 30 words with their counts

```python
#----------------------------------------------
# Your code starts here
#   Please add comments or text cells in between to explain the general idea of each block of the code.
#   Please feel free to add more cells below this cell if necessary

# Compute a collection of all words from all tweets
status_texts = [ status['text'] 
                 for status in statuses ]
words = [ w 
          for t in status_texts 
              for w in t.split() ]


# Explore the first 5 results of tweets/users mentioned/hashtags 
# and the words being used in all tweets
#print json.dumps(status_texts[0:5], indent=5)
#print json.dumps(words[0:10], indent=5)


# Creating a basic frequency distribution from the words in tweets
from collections import Counter

for item in [words]:
    c = Counter(item)
    print c.most_common()[:30]  # top30
    print
    

# Plot a table of the top 30 words with their counts
from prettytable import PrettyTable

for label, data in (('Word', words),):  # ('Screen Name', screen_names), ('Hashtag', hashtags)
    pt = PrettyTable(field_names=[label, 'Count']) 
    c = Counter(data)
    [ pt.add_row(kv) for kv in c.most_common()[:30] ]
    pt.align[label], pt.align['Count'] = 'l', 'r' # Set column alignment
    print pt
```

**2. Find the most popular tweets in your collection of tweets**

Please plot a table of the top 10 tweets that are the most popular among your collection, i.e., the tweets with the largest number of retweet counts.


```python
#----------------------------------------------
# Your code starts here
#   Please add comments or text cells in between to explain the general idea of each block of the code.
#   Please feel free to add more cells below this cell if necessary

retweets = [
            # Store out a tuple of these three values ...
            (status['retweet_count'], 
             status['retweeted_status']['user']['screen_name'],
             status['text']) 
            
            # ... for each status ...
            for status in statuses 
            
            # ... so long as the status meets this condition.
                if status.has_key('retweeted_status')
           ]

# Slice off the first 10 from the sorted results and display each item in the tuple

pt = PrettyTable(field_names=['Count', 'Screen Name', 'Text'])
[ pt.add_row(row) for row in sorted(retweets, reverse=True)[:10] ]
pt.max_width['Text'] = 50
pt.align= 'l'
print pt

```

**3. Find the most popular Tweet Entities in your collection of tweets**

Please plot a table of the top 10 hashtags, top 10 user mentions that are the most popular in your collection of tweets.

```python
#----------------------------------------------
# Your code starts here
#   Please add comments or text cells in between to explain the general idea of each block of the code.
#   Please feel free to add more cells below this cell if necessary

# Compute a collection of all users mentioned '@' from all tweets
screen_names = [ user_mention['screen_name'] 
                 for status in statuses
                     for user_mention in status['entities']['user_mentions'] ]

# Compute a collection of all hashtags from all tweets
hashtags = [ hashtag['text'] 
             for status in statuses
                 for hashtag in status['entities']['hashtags'] ]


# Creating a basic frequency distribution from the screen_names and hashtags in tweets
from collections import Counter
for item in [screen_names, hashtags]:
    c = Counter(item)
    print c.most_common()[:10]  # top10
    print
    

# Plot a table of the top 10 user mentions and hashtags with their counts
from prettytable import PrettyTable

for label, data in (('User Mention', screen_names), 
                    ('Hashtag', hashtags)):
    pt = PrettyTable(field_names=[label, 'Count']) 
    c = Counter(data)
    [ pt.add_row(kv) for kv in c.most_common()[:10] ]
    pt.align[label], pt.align['Count'] = 'l', 'r' # Set column alignment
    print pt

```

* Compute the mutual friends within the two groups, i.e., the users who are in both friend list and follower list, plot their ID numbers and screen names in a table

```python
#----------------------------------------------
# Your code starts here
#   Please add comments or text cells in between to explain the general idea of each block of the code.
#   Please feel free to add more cells below this cell if necessary

mutual_id = list( set(friends_id) & set(followers_id) )
number_mutual = len(mutual_id)
print 'Numbers of mutual friends:', number_mutual

ids = str(mutual_id).strip('[]')
mutual_name = [user['name']
                for user in twitter_api.users.lookup(user_id=ids)]

mutual_info = [(mutual_id[i],mutual_name[i]) for i in range(number_mutual)]

pt = PrettyTable(field_names=['ID number', 'Screen Name'])
[ pt.add_row(row) for row in mutual_info ]
pt.align= 'l'
print pt
```
